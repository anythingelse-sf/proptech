```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
## Import the Data from the resources folder 
```

```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(Path('./Resources/sfo_neighborhoods_census_data.csv'))

# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
sfo_data_df.tail()
```

```python
## Calculate and Plot the Housing Units per Year
```

```python
# Sort out the housing units by year
housing_units_by_year = (sfo_data_df[['year','housing_units']])

# Review the DataFrame
display(housing_units_by_year)
```

```python
# Use groupby and mean to create a numerical aggregation that groups data by the year then averages the results
avg_housing_units_by_year=housing_units_by_year.groupby('year').mean()
display(avg_housing_units_by_year)
```

```python
# Create a visual aggregation explore the housing units by year using hvplot
avg_housing_units_by_year.hvplot(kind='bar',title='Average Housing Units Available by Year', ylim=(350000,390000)).opts(yformatter='%.0f')
```

```python
## Calculate and Plot the Average Sale Prices per Square Foot
```

```python
# Sort out sale_price_sqr_foot and gross_rent by year
prices_square_foot_by_year = (sfo_data_df[['year','sale_price_sqr_foot','gross_rent']])

# Review the resulting DataFrame
prices_square_foot_by_year
```

```python
# Use groupby and mean to create a numerical aggregation that groups data by the year then averages the results
avg_prices_square_foot_by_year=(prices_square_foot_by_year.groupby('year').mean())
display(avg_prices_square_foot_by_year)
```

```python
# Plot prices_square_foot_by_year. 
avg_prices_square_foot_by_year.hvplot(title='Compare the Avg Price to Buy vs Avg Gross Rent')
```

```python
## Build an Interactive Neighborhood Map
```

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(Path('./Resources/neighborhoods_coordinates.csv'))

# Review the DataFrame
neighborhood_locations_df.set_index('Neighborhood')
```

```python
# Calculate the mean values for each neighborhood
all_neighborhood_info_df = sfo_data_df.groupby('neighborhood').mean()

# Review the resulting DataFrame
all_neighborhood_info_df
```

```python

neighborhood_locations_df=neighborhood_locations_df.rename(columns={'Neighborhood':'neighborhood'})
neighborhood_locations_df=neighborhood_locations_df.set_index('neighborhood')
```

```python
# Using the Pandas `concat` function, join the 
# neighborhood_locations_df and the all_neighborhood_info_df DataFrame
# The axis of the concatenation is "columns".
# The concat function will automatially combine columns with
# identical information, while keeping the additional columns.
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False,
)
all_neighborhoods_df=all_neighborhoods_df.dropna()
# Review the resulting DataFrame
all_neighborhoods_df
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
all_neighborhoods_df = all_neighborhoods_df.reset_index().dropna()

# Rename the "index" column as "Neighborhood" for use in the Visualization
all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"})

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
# # Create a plot to analyze neighborhood info
# * Set the `geo` parameter to True.
# * Set the `size` parameter to “sale_price_sqr_foot”.
# * Set the `color` parameter to “gross_rent”.
# * Set the `frame_width` parameter to 700.
# * Set the `frame_height` parameter to 500.
# * Include a descriptive title.
# # YOUR CODE HERE
```

```python
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size='sale_price_sqr_foot',
    color='gross_rent',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title= 'Average by Neighborhood',
    scale=.5
    )
```
